```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math as m

%matplotlib inline

def det(A):
    det = np.linalg.slogdet(A)
    return det

N = 2**np.arange(6, 12)

times = []

for n in N:
    A = np.random.rand(n, n)
    t = %timeit -q -n4 -r1 -o det(A)
    times.append(t.best)

    
plt.loglog(N, times, marker='o', label='linear search')
plt.xlabel('$n$')
plt.ylabel('$t$ (s)')
plt.loglog(N, 1e-6*N**2.373, label='$O(n^2)$')
plt.legend(loc=0)
plt.show()
```
