```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from ipywidgets import interact

# Plot interactive graph of mixed sines.

def plot_1(ω=1, α=0):
    x = np.linspace(-4*np.pi, 4*np.pi, 100)
    plt.xlabel('$x$')
    plt.ylabel('$f$')
    plt.title(r"$\alpha$ = {}, $\omega$ = {}".format(α, ω))
    plt.plot(x, (np.exp(x/10)*np.sin(α*x)*np.cos(ω*x)))
    plt.show()
    
interact(plot_1, ω=(0, 2, 0.1), α=(0, 2, 0.1));




# Plot a tending sin graph.

x = np.linspace(-6*np.pi, 6*np.pi, 100)
f = (np.sin(x))/x
g = 1/abs(x)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title("Simple plot of $x$ vs $f$")
plt.plot(x,f)
plt.plot(x,g, '-r')
plt.ylim(-1, 1)
plt.show()


# Interactive pie chart of population.


def pie(ω=2011, αa=0, αb=0, αc=0, αd=0, αe=0):
    district = ('Cambridge City', 'East Cambridgeshire', 'Fenland', 'Huntingdonshire', 'South Cambridgeshire')
    population = ((123900*(1+αa/100)**(ω-2011)), (83800*(1+αb/100)**(ω-2011)), (95300*(1+αc/100)**(ω-2011)), (169500*(1+αd/100)**(ω-2011)), (148800*(1+αe/100)**(ω-2011)))
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red']
    explode = (0.1, 0.0, 0, 0, 0)  
    plt.pie(population, explode=explode, labels=district, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
    plt.title(r"Population in Cambridge Districts in {}".format(ω))
    plt.show
    
    
interact(pie, ω=(2011, 2021, 1), αa=(0, 10, 1), αb=(0, 10, 1), αc=(0, 10, 1), αd=(0, 10, 1), αe=(0, 10, 1));


# Plotting crime.

import json
import requests
import pprint


long, lat = 52.205277, 0.119117
year_month = '2019-09'
url = "https://data.police.uk/api/crimes-street/all-crime?lat={}&lng={}&date={}".format(long, lat, year_month)
r = requests.get(url)
crime_data = r.json()

categories_freq = {}

for report in crime_data:
    category = report['category']
    if category in categories_freq:
        categories_freq[category] += 1
    else:
        categories_freq[category] = 1
    
categories = []
for key in categories_freq:
    categories.append(key)
frequency = []
for key in categories_freq:
    frequency.append(categories_freq[key])


x_pos = np.arange(len(categories_freq))
plt.bar(x_pos, frequency, align='center');
plt.xticks(x_pos, categories, rotation=90)
plt.xlabel('Catergories')
plt.ylabel('Frequency')
plt.show()




```
